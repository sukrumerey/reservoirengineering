# Introduction

## Suggested Books for Reservoir Engineering
- Dake, L.P. (1978). Fundamentals of Reservoir Engineering. Elsevier.
- Society of Petroleum Engineers (SPE). (2006). Petroleum Engineering Handbook. SPE.
- Society of Petroleum Evaluation Engineers (SPEE). Monograph 4: Estimating Developed Reserves in Unconventional Reservoirs.
- Dake, L.P. (2001). The Practice of Reservoir Engineering. Elsevier.
- McCain, W.D. (1990). Modern Reservoir Engineering: A Simulation Approach. PennWell Books.
- Lee, J., & Samuel, R. (2015). Shale Reservoir Engineering: Theory and Practice. Gulf Professional Publishing.
- McCain, W.D., Jr. (1990). The Properties of Petroleum Fluids. PennWell Books.
- Sanni, M. (2018). Petroleum Engineering: Principles, Calculations, and Workflows. John Wiley & Sons.
